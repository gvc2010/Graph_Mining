```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Carregar a matriz de adjacÃªncias do arquivo CSV
matriz_adj1 = pd.read_csv('Grafo3.csv', delimiter=',', header=None)

# Criar o grafo G a partir da matriz de adjacÃªncias
G3 = nx.from_pandas_adjacency(matriz_adj1)

# Verificar se o grafo Ã© direcionado
direcionado = nx.is_directed(G3)
print('O grafo G3 Ã© direcionado:', direcionado)

# Desenhar o grafo G na tela com dimensÃµes 12 x 7
plt.figure(figsize=(12, 7))
nx.draw(G3, with_labels=True, node_size=300, node_color='skyblue', font_size=10, font_color='black', arrowsize=10)
plt.title('Grafo G3')
plt.show()

# ExtraÃ§Ã£o da quantidade de vÃ©rtices e de arestas do grafo G
quantidade_vertices = G3.order()
quantidade_arestas = G3.size()
print('Quantidade de vÃ©rtices do grafo G3:', quantidade_vertices)
print('Quantidade de arestas do grafo G3:', quantidade_arestas)

# ExtraÃ§Ã£o do grau do vÃ©rtice 33 e diferenÃ§a em relaÃ§Ã£o ao grau mÃ©dio da rede
grau_vertice_33 = G3.degree(33)
graus_G3 = dict(G3.degree())
grau_medio_G3 = np.mean(list(graus_G3.values()))
diferenca_grau = grau_vertice_33 - grau_medio_G3
print('Grau do vÃ©rtice 33 do grafo G3:', grau_vertice_33)
print('DiferenÃ§a do grau do vÃ©rtice 33 em relaÃ§Ã£o ao grau mÃ©dio da rede:', diferenca_grau)

# ExtraÃ§Ã£o do conjunto de arestas que se conectam ao vÃ©rtice 33
arestas_vertice_33 = list(G3.edges(33))
print('Conjunto de arestas que se conectam ao vÃ©rtice 33:', arestas_vertice_33)

# Apresentar a distribuiÃ§Ã£o de graus do grafo G em uma figura com dimensÃµes 10 x 5
plt.figure(figsize=(10, 5))
plt.hist(list(graus_G3.values()), bins=20, color='skyblue', edgecolor='black')
plt.title('DistribuiÃ§Ã£o de graus do Grafo G3')
plt.xlabel('Grau')
plt.ylabel('Quantidade')
plt.grid(True)
plt.show()
```
